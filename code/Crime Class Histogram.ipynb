```python
import pandas as pd  
import numpy as np
%matplotlib inline 
import matplotlib.pyplot as plt 
df = pd.read_csv('master_iu_crime.csv', delimiter=',')
crime = list(df['crime_class'])
crime_dic={}
##plt.xlabel('Crime Type',fontsize=20)
##plt.ylabel('Total number of Crime',fontsize=20)
##plt.title("Frequency of type of crime on Campus from 13-15",fontsize=20)
for i in range(len(crime)-1):
    if crime[i] not in crime_dic:
        crime_dic[crime[i]] = 1
    else:
        crime_dic[crime[i]]+=1
print(crime_dic)
plt.xlabel('Crime Classes',fontsize=20)
plt.ylabel('Frequency',fontsize=20)
plt.title("Master IU Crime",fontsize=20)
plt.bar(range(len(crime_dic)), crime_dic.values(), align="center",color='yellow')
plt.xticks(range(len(crime_dic)), crime_dic.keys())
plt.grid()
fig = plt.gcf()
fig.set_size_inches(34.5, 15.5)
fig.savefig('masteriucrime.png')
```
