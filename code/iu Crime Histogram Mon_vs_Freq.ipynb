```python
import pandas as pd  
import numpy as np
%matplotlib inline 
import time
import calendar
import matplotlib.pyplot as plt 

df = pd.read_csv('master_iu_crime.csv', delimiter=',')
crime_month = list(df['date'])
crime = list(df['crime_class'])
crime_dic={}
##plt.xlabel('Crime Type',fontsize=20)
##plt.ylabel('Total number of Crime',fontsize=20)
##plt.title("Frequency of type of crime on Campus from 13-15",fontsize=20)
for i in range(len(crime_month)-1):
    if crime_month[i] not in crime_dic:
        crime_dic[crime_month[i]] = 1
    else:
        crime_dic[crime_month[i]]+=1
print(crime_dic)
plt.xlabel('Monthes',fontsize=15)
plt.ylabel('Frequency',fontsize=15)
plt.title("IU Campus Crime: Monthes vs Frequency in 2013-2015",fontsize=15)
plt.bar(range(len(crime_dic)), crime_dic.values(), align="center",color='yellow')
plt.xticks(range(len(crime_dic)), crime_dic.keys())
plt.grid()
fig = plt.gcf()
fig.set_size_inches(12.5, 8.5)
fig.savefig('iucrime.svg')
```
