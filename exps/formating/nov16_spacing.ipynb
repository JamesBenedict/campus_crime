```python
class Formater():
    """ class for formatting the 2016 txt from pdf data in a more consistent manner """

    def __init__(self, content):
        self.info = []
        self.crime_list = []
        
        self.spacing(content)
        self.removeHead()
        self.chunker(self.info)
     
        
    def spacing(self, content):
        line_itr = iter(content.splitlines())
        for line in line_itr:
            if len(line) > 0:
                self.info.append(line)
                
    def removeHead(self):
        if self.info[0] == 'Indiana University, Bloomington':
            self.info.pop(0)
        if self.info[0] == 'Police Department':
            self.info.pop(0)
        if self.info[0] == 'Daily crime log report':
            self.info.pop(0)
        if 'from' in self.info[0]:
            self.info.pop(0)
            
   
    def chunker(self, content):
        chunk_start = []
        chunk_end = []
        chunks = {}
        crime_list = []
        for i in range(len(content)):
            if 'date reported' in content[i]:
                chunk_start.append(i)
            elif 'modified date' in content[i]:
                chunk_end.append(i+1)
                
        for i in range(len(chunk_start)):
            chunks[chunk_start[i]] = chunk_end[i]
            
        for key, value in chunks.items():
            
            self.crime_list.append(content[key:value])
#         for crime in crime_list:
#             print(crime)

#     def reportNumSwitch(self, content):
#         for line in 
r = Formater(open(fname).read())
pprint(r.crime_list)
# pprint(r.crime_list)
```
