```python
%load_ext autoreload
%autoreload 2
import os
import glob
import random
import re
from datetime import *
from pprint import pprint
```

```python
path = '../../data/scrapped/text/2016'
files = glob.glob(os.path.join(path, '*'))
fname = random.choice(files)

print(open(fname).read())
```

```python
class Formater():
    """ class for formatting the 2016 txt from pdf data in a more consistent manner """

    def __init__(self, content):
        self.info = []
        self.chunks = []
        self.spacing(content)
        self.removeHead(content)
        
    def spacing(self, content):
        line_itr = iter(content.splitlines())
        for line in line_itr:
            if len(line) > 0:
                self.info.append(line)
                
    def removeHead(self, content):
        if self.info[0] == 'Indiana University, Bloomington':
            self.info.pop(0)
        if self.info[0] == 'Police Department':
            self.info.pop(0)
        if self.info[0] == 'Daily crime log report':
            self.info.pop(0)
        if 'from' in self.info[0]:
            self.info.pop(0)
            
#     def crime_chunker(self, content):
#         self.chunks.append(re.findall('d', self.info, re.DOTALL))
                
r = Formater(open(fname).read())
pprint(r.info)
```
